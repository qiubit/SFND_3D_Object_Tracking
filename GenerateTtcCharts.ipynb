```python
from glob import glob
from os import path

import matplotlib.pyplot as plt
import numpy as np
```

```python
DETECTORS = ['SHITOMASI', 'HARRIS', 'AKAZE', 'BRISK', 'FAST', 'ORB', 'SIFT']
DESCRIPTORS = ['BRISK', 'BRIEF', 'ORB', 'FREAK', 'AKAZE', 'SIFT']
MEASUREMENTS_DIR = 'dat/ttc_cam_measurements'
CHART_DIR = 'dat/ttc_cam_charts'
```

```python
detector = DETECTORS[0]
ttcMeasurements = []

plt.figure(figsize=(12, 8))
plt.title(detector)
plt.xticks(np.arange(18))
plt.xlabel('Number of image in image sequence')
plt.ylabel('Estimated TTC (s)')

plottedDescriptors = []

for descriptor in DESCRIPTORS:
    measurementsFile = path.join(MEASUREMENTS_DIR, '{}_{}.out'.format(detector, descriptor))

    if not path.exists(measurementsFile):
        continue
    
    plottedDescriptors.append(descriptor)

    with open(measurementsFile) as f:
        lines = f.readlines()
        measurements = [float(m[:-1]) for m in lines]
        ttcMeasurements.append(measurements)
        plt.plot(measurements)
        
plt.legend(plottedDescriptors)

plt.savefig(path.join(CHART_DIR, '{}.png'.format(detector)))
```
